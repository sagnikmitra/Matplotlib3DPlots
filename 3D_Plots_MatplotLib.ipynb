```
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
z = inico['C']
x = inico['A']
y = inico['B']
 
# Creating figure
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
 
# Creating plot
ax.scatter3D(x, y, z, color = "blue")
plt.title("FCC Initial Co-ordination")
plt.xlabel('X-axis')  
plt.ylabel('Y-axis') 
# Setting Z Axis Label with plt.axes
ax.set_zlabel('Z-Axis')
# show plot
plt.show()
```
